### Create a function which will take two string(name of two teams) as input. Show win Loss record between them and player getting most player of the match award in matches between these two teams.

```python
def win_loss_record_and_MOM(ipl,team1,team2):
    f1 = (ipl.Team1 == team1) & (ipl.Team2 == team2)
    f2 = (ipl.Team1 == team2) & (ipl.Team2 == team1)
    pom = ipl[f1|f2].Player_of_Match.value_counts().index[0]
    win_dict = ipl[f1|f2].WinningTeam.value_counts().to_dict()
    if team1 not in win_dict:
        return f"In matches between {team1} and {team2}, {team1} wins {0} matches and {team2} wins {win_dict[team2]} matches and most player of the match award win by {pom}"
    elif team2 not in win_dict:
        return f"In matches between {team1} and {team2}, {team1} wins {win_dict[team1]} matches and {team2} wins {0} matches and most player of the match award win by {pom}"
    else:
        return f"In matches between {team1} and {team2}, {team1} wins {win_dict[team1]} matches and {team2} wins {win_dict[team2]} matches and most player of the match award win by {pom}"
```

```python
win_loss_record_and_MOM(ipl,"Chennai Super Kings","Kolkata Knight Riders")
```
