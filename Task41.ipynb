### Q.2:The average test score in a certain statistics class was 74 with a standard deviation of 8. There are 2000 students in this class. Use the emperical rule to answer the following questions:

```python
mean = 74
std = 8 
n = 2000
```

#### a. What percentage of students scored less than 58?

```python
x = 58
def z_score(x,mean,std):
    z = (x - mean)/std
    return 'Value of z_score is {} for x = {}'.format(z,x)
print(z_score(58,74,8))
# As per emperical rule ie 68-95-97, -2std to + 2std probability is 95.45% and graph is symmetrical so we want graph under the curve before -2std 
# to calculate it we have total % 100 , for -2std to + 2std  95.45% so 100 - 95.45 = 4.55 , as graph is symmetrical so area under the curve beyond 
# 2std and -2std is same we want only are beyond -2std hence 
print(f'% student scored less than 58 is {4.55/2}')
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate some normally distributed data
mu = 0
sigma = 1
data = np.random.normal(mu, sigma, size=10000)

# Compute the mean and standard deviation of the data
mean = np.mean(data)
std = np.std(data)

# Compute the percentages for each standard deviation
pct_0 = round(((np.sum((data >= mean - std) & (data <= mean + std))) / len(data)) * 100, 2)
pct_1 = round(((np.sum((data >= mean - 2*std) & (data <= mean + 2*std))) / len(data)) * 100, 2)
pct_2 = round(((np.sum((data >= mean - 3*std) & (data <= mean + 3*std))) / len(data)) * 100, 2)

percentages = {
    1: pct_0,
    2: pct_1,
    3: pct_2
}

fig = plt.figure(figsize=(16, 9))

# Plot the histogram of the data
plt.hist(data, bins=60, density=True, alpha=0.5)

#Plot the empirical rule lines with markers and percentages
for i in range(-3, 4):
    x = mean + i*std
    y = 1/(np.sqrt(2*np.pi)*std)*np.exp(-(x-mean)**2/(2*std**2))
    plt.plot([x, x], [0, y], linestyle='--', color='r', alpha=0.7)
    if abs(i) in percentages:
        plt.annotate(
            f"{i} std\n({percentages[abs(i)]}%)",
            (x, y),
            xytext=(x - 0.2, y + 0.01),
            fontsize=8,
            color='r'
        )
# Add labels and a title to the plot
plt.xlabel('x')
plt.ylabel('Frequency')
plt.title('Normal Distribution with Empirical Rule')

# Show the plot
plt.show()
```

### `Q-3:` Normally distributed IQ scores have a mean of 100 and a standard deviation of 15. Use the standard z-table to answer the following questions:

What is the probability of randomly selecting someone with an IQ score that is 

    (a) less than 80? 
    (b) greater than 136? 
    (c) between 95 and 110?
    (d) What IQ score corresponds to the 90th percentile? 
    (e) The middle 30% of IQs fall between what two values?

```python
mean = 100
std = 15
print(f'mean is {mean}')
print(f'std is {std}')
```

#### IQ less than 80 ?

```python
x = 80
def z_score(x,mean,std):
    z = (x - mean)/std
    return 'Value of z_score is {} for x = {}'.format(z,x)
print(z_score(x,mean,std))
# values from z table 0.091800 
#Using the z-table, the probability of a randomly selected person having an IQ score less than 80 is approximately 0.091800 or 9.18%.
print(f"Probability of IQ value having less than 80 is {9.81}%")
```

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Mean and standard deviation
mean = 100
std = 15

# Create data points for x-axis
x = np.linspace(mean - 3*std, mean + 3*std, 100)

# Calculate the corresponding y-values (PDF)
y = 1/(std * np.sqrt(2 * np.pi)) * np.exp(-(x - mean)**2 / (2 * std**2))

# Initialize the plot
fig, ax = plt.subplots(figsize=(10,6))

# Plot the normal distribution curve
ax.plot(x, y, color='black', linewidth=2)

# Shade the areas corresponding to 1, 2, and 3 standard deviations from the mean
ax.fill_between(x, y, where=(x >= mean - std) & (x <= mean + std), alpha=0.3, color='blue')
ax.fill_between(x, y, where=(x >= mean - 2*std) & (x <= mean + 2*std), alpha=0.3, color='orange')
ax.fill_between(x, y, where=(x >= mean - 3*std) & (x <= mean + 3*std), alpha=0.3, color='green')

# Add a vertical line to indicate the mean
ax.axvline(x=mean, color='green', linestyle='--', linewidth=2)

# Set axis labels and title
ax.set_xlabel('IQ Scores')
ax.set_ylabel('Probability Density')
ax.set_title('Normal Distribution of IQ Scores')

# Display the plot
plt.show()
```

```python
import pandas as pd
import scipy.stats as st

# Define the range of z-scores to be included in the table
z_range_p_ve = np.arange(0.0, 3.5, 0.1)
col = np.arange(0.00, 1.0, 0.1)/10
# Create a pandas dataframe to store the z-score table
z_table = pd.DataFrame(index=z_range_p_ve, columns=col)

# Fill the dataframe with the corresponding probability values
for i in z_range_p_ve:
    for j in col:
        z_table.loc[i,j] = round(st.norm.cdf(j+i), 4)
print("**+ve z- score table**")
# Print the z-score table
z_table
```

#### b. greater than 136? 

```python
x = 136
def z_score(x,mean,std):
    z = (x - mean)/std
    return 'Value of z_score is {} for x = {}'.format(z,x)
print(z_score(x,mean,std))
# values from z table 0.9918 
#Using the z-table, the probability of a randomly selected person having an IQ score greater than 136 is approximately 1 - 0.9918 or 0.82%.
print(f"Probability of IQ value having greater than 136 is {0.82}%")
```

#### c. What is the probability of randomly selecting someone with an IQ score between 95 and 110?

```python
x1 = 95
x2 = 110

def z_score(x,mean,std):
    z = (x - mean)/std
    return 'Value of z_score is {} for x = {}'.format(z,x)
print(z_score(x1,mean,std))
print(z_score(x2,mean,std))
# value from z table for x1 is 0.366900
# value from z table for x2 is 0.748600
print("probability of randomly selecting someone with an IQ score between 95 and 110 {}%".format((0.748600-0.366900)*100))
```

#### d. What IQ score corresponds to the 90th percentile?

```python
#The 90th percentile corresponds to a z-score of 1.28 (from the z-table) We can use this z-score to find the corresponding IQ score:
z = 1.28 
mean = 100
std = 15
def find_x(z,mean,std):
    x = (z*std) + mean
    return 'Value of x is {}'.format(x)
print(f"Therefore, the IQ score corresponding to the 90th percentile is approximately {find_x(z,mean,std)}")
```

#### e. The middle 30% of IQs fall between what two values?

The middle 30% of IQs corresponds to the area between the 35th and 65th percentiles (since the total area under the curve is 100% and we want the middle 30%)

The z-scores corresponding to the 35th and 65th percentiles are approximately -0.39 and 0.39, respectively (from the z-table)

We can use these z-scores to find the corresponding IQ scores:

IQ1 = -0.39(15) + 100 = 94.65

IQ2 = 0.39(15) + 100 = 105.85

Therefore, the middle 30% of IQs fall between approximately 94.65 and 105.85.
